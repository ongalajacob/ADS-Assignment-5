```python

sns.set_context("talk", font_scale=1.1)
plt.figure(figsize=(10,6))
sns.scatterplot(x="Age", 
                y="Fare",
                size="Pclass",hue="Survived" ,         
                data=df[df.Fare<=100])
plt.legend(bbox_to_anchor=(1.01, 1),borderaxespad=0)
plt.xlabel("Age")
plt.ylabel("fare")
plt.tight_layout()
```
